# Note 
* inertia_ returns sum of square distance within cluster - how close values are to cluster centroid
* Larger the value - more spread out the cluster is 

```python
kmeans.inertia_
```

```python
kmeans.score(data)
```

```python
SSE = []
```

# Note
* We need to find out the optimal number of clusters that a particular dataset should have to get the least possible 'inertia' and so we use a loop and evaluate clustering and store the value of inertia in an array called 'SSE' from which we select the best possible clustering

```python
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(data)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
```

```python
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_scaled = scaler.fit_transform(data)
```

```python
pd.DataFrame(data_scaled).describe()
```

```python
SSE_scaled = []
```

```python
for cluster in range(1,20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(data_scaled)
    SSE_scaled.append(kmeans.inertia_)
```

```python
frame_scaled = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE_scaled})
plt.plot(frame_scaled['Cluster'], frame_scaled['SSE'], marker='o')
plt.xlabel("Clusters")
plt.ylabel("SSE")
```
